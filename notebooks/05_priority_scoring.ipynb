# Demo: Priority Scoring

This notebook demonstrates `create_priority_table` with a small demand table, supply table, and district geography sample.

```python
from pathlib import Path
import sys

def add_app_python_src_to_path():
    for base in [Path.cwd(), *Path.cwd().parents]:
        candidate = base / "dais-health-access" / "python" / "src"
        if candidate.exists():
            sys.path.insert(0, str(candidate))
            return candidate
    raise FileNotFoundError("Could not find dais-health-access/python/src from the current working directory")

APP_SRC = add_app_python_src_to_path()
```

```python
import pandas as pd
from facility_prioritization.priority_scoring import create_priority_table

demand_df = pd.DataFrame(
    [
        {"district_name": "Alpha", "state_ut": "State 1", "treatment": "cardiology", "treatment_score": 2.1},
        {"district_name": "Alpha", "state_ut": "State 1", "treatment": "oncology", "treatment_score": 1.1},
        {"district_name": "Beta", "state_ut": "State 1", "treatment": "cardiology", "treatment_score": 0.7},
        {"district_name": "Gamma", "state_ut": "State 2", "treatment": "pediatrics", "treatment_score": 1.8},
    ]
)

supply_df = pd.DataFrame(
    [
        {"name": "Facility A", "latitude": 28.61, "longitude": 77.20, "top_treatments_offered": ["cardiology", "pediatrics"]},
        {"name": "Facility B", "latitude": 19.07, "longitude": 72.88, "top_treatments_offered": ["oncology"]},
        {"name": "Facility C", "latitude": 13.08, "longitude": 80.27, "top_treatments_offered": ["pediatrics", "radiology"]},
    ]
)

geo_reference_df = pd.DataFrame(
    [
        {"district": "Alpha", "latitude": 28.70, "longitude": 77.10, "statename": "State 1"},
        {"district": "Beta", "latitude": 19.20, "longitude": 72.90, "statename": "State 1"},
        {"district": "Gamma", "latitude": 13.00, "longitude": 80.20, "statename": "State 2"},
    ]
)

priority_table, warnings = create_priority_table(demand_df, supply_df, geo_reference_df)

print("Warnings:")
for warning in warnings:
    print(f"- {warning}")

priority_table
```
