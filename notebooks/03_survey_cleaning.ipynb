# Demo: Survey Cleaning

This notebook demonstrates `clean_survey_data` with a small district-level survey sample.

```python
from pathlib import Path
import sys

def add_app_python_src_to_path():
    for base in [Path.cwd(), *Path.cwd().parents]:
        candidate = base / "dais-health-access" / "python" / "src"
        if candidate.exists():
            sys.path.insert(0, str(candidate))
            return candidate
    raise FileNotFoundError("Could not find dais-health-access/python/src from the current working directory")

APP_SRC = add_app_python_src_to_path()
```

```python
import pandas as pd
from facility_prioritization.config import load_config
from facility_prioritization.survey_processing import clean_survey_data

config = load_config(APP_SRC.parent.parent / "config" / "config.yaml")

raw_survey = pd.DataFrame(
    [
        {"district_name": "Alpha", "state_ut": "State 1", "asthma_pct": "10.5", "maternal_care_gap": "4.2", "bad_text": "n/a"},
        {"district_name": "Beta", "state_ut": "State 1", "asthma_pct": "13.1", "maternal_care_gap": "8.9", "bad_text": "missing"},
        {"district_name": "Gamma", "state_ut": "State 2", "asthma_pct": "7.2", "maternal_care_gap": "5.4", "bad_text": "unknown"},
    ]
)

cleaned_survey, warnings = clean_survey_data(raw_survey, config=config)

print("Warnings:")
for warning in warnings:
    print(f"- {warning}")

cleaned_survey
```
